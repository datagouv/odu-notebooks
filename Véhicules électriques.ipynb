# Défi open data : Véhicules électriques

En France, le secteur des transports est responsable de 38% des émissions de dioxyde de carbone (CO2). Pour opérer une transition vers une mobilité plus durable, le Gouvernement s’est engagé dans **le développement de la mobilité électrique**. Ainsi, [la loi d’orientation des mobilités](https://www.ecologie.gouv.fr/loi-dorientation-des-mobilites#:~:text=La%20loi%20d%27orientation%20des%20mobilit%C3%A9s%20a%20%C3%A9t%C3%A9%20publi%C3%A9e%20au,moins%20co%C3%BBteux%20et%20plus%20propres.) fixe-t-elle comme objectif la fin de la vente de voitures particulières et de véhicules utilitaires légers neufs utilisant des énergies fossiles d’ici 2040.

Depuis 2020, les immatriculations de véhicules électriques connaissent une forte progression, le nombre de ventes ayant bondi de 128% en 2020 et de 63% en 2021.

Ce déploiement doit nécessairement **s’accompagner d’un développement des infrastructures de recharge**.

Pour que le modèle du véhicule électrique soit viable, il est nécessaire de disposer d’une borne de recharge publique pour 10 voitures (Directive AFI). La France a largement dépassé ces préconisations. Mais si les objectifs de couverture territoriale homogène demeurent pertinents, le développement de la mobilité électrique passe désormais par **une adéquation des infrastructures de recharge aux besoins des usagers**.

**La problématique**

Ce projet vous propose ainsi de répondre aux deux questions suivantes :
* **Comment le nombre d’infrastructures de recharge évolue-t-il en France ? Comparer cette évolution avec celle du parc de véhicules électriques et les objectifs pour 2030 ainsi que d’autres indicateurs qui vous semblent pertinents.**
* **La répartition des infrastructures de recharge sur le territoire correspond-elle à une logique de distribution de la population et de trafic routier ?**

Beaucoup de données ouvertes sont mises à disposition pour vous permettre de répondre à ce sujet. Plusieurs jeux de données ont déjà été chargés dans cet environnement ([cf. liste](https://defis.data.gouv.fr/defis/65a923a083cf5f728c9934b3/)) et vous pouvez bien évidemment en ajouter d’autres selon vos besoins.

# Exemple de réutilisation : Visualiser l'évolution du nombre de véhicules électriques en France et la densité des infrastructures de recharge

```python
import pandas as pd
import plotly.express as px
```

```python
voit_comm = load_table_from_resource_id(
    '4e4fccdb-6acb-4e31-8b2d-cb170f639f1a',
    sep=';',
    dtype={
        'codgeo': str,
        'epci': str,
    }
)
voit_comm
```

```python
fig = px.bar(
    voit_comm.groupby('date_arrete')['nb_vp_rechargeables_el'].sum(),
    title='Evolution du nombre de véhicules électriques en France',
)
fig.show(
#     renderer='png'
)
```

```python
irve = load_table_from_resource_id(
    'eb76d20a-8501-400e-b336-d85724de5435',
    dtype={
        "siren_amenageur": str,
        "code_insee_commune": str,
        "consolidated_code_postal": str,
    }
)
irve
```

```python
fig = px.density_mapbox(
    irve,
    lat="consolidated_latitude",
    lon="consolidated_longitude",
    z=[1 for k in range(len(irve))],
    radius=2,
    mapbox_style="open-street-map",
    center={"lat": irve["consolidated_latitude"].mean(), "lon": irve["consolidated_longitude"].mean()},
    zoom=4,
    color_continuous_scale='RdYlGn_r'
)
fig.show(
#     renderer='png'
)
```
