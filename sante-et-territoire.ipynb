# Défi open data : Santé et territoire

La crise sanitaire de 2020 a mis un coup de projecteur sur **les politiques de santé en France**. La santé, définie par l’[Organisation Mondiale de la Santé](https://www.who.int/fr/about/frequently-asked-questions) (OMS) comme un état de bien-être physique, mental et social, ne se limite pas à l'absence de maladie.

La santé est aussi influencée par divers facteurs, tels que les pratiques sociales, le groupe social et le genre. Par exemple, les maladies chroniques comme les diabètes ou les cancers sont plus fréquentes chez les personnes à faibles revenus ([source](https://drees.solidarites-sante.gouv.fr/sites/default/files/2023-02/Annexes5.pdf)). C’est ce qu’on appelle [**les déterminants sociaux de la santé** ](https://www.canada.ca/fr/sante-publique/services/promotion-sante/sante-population/est-determine-sante.html)(*cf. “Pour en savoir plus”*). 

Pour améliorer la santé d’une population, les pouvoirs publics mettent en œuvre des actions de [**santé publique**](https://geoconfluences.ens-lyon.fr/glossaire/sante-sante-publique) pour à la fois identifier les besoins de santé d’une population, mais aussi organiser l'ensemble des services. Les pouvoirs publics peuvent par exemple soutenir l’offre de soins (hôpitaux, installation des professionnels de santé, développement des [maisons pluriprofessionnelles de santé](https://fr.wikipedia.org/wiki/Maison_de_sant%C3%A9_pluridisciplinaire), …). Ils peuvent aussi **agir sur d’autres facteurs** qui ont une incidence sur la santé : le logement, l’environnement (qualité de l’air), l’éducation, le travail… Les **actions de prévention** entrent également dans le champ de la santé publique. 

Ce projet vous propose de réaliser un **diagnostic territorial de santé publique**, afin de guider les collectivités locales dans la mise en place d’actions de prévention. Il se concentrera sur **l’offre de soin, les déterminants sociaux de santé, et l’identification des populations vulnérables** (*voir section “Pour en savoir plus”*).

### La problématique

# Exemple de réutilisation : Répartition des établissements d'hébergement pour personnes âgées par département et comparaison avec la population

```python
import pandas as pd
import geopandas as gpd

pd.set_option("display.max_columns", None)
```

```python
cat = "Etablissements d'Hébergement pour Personnes Âgées"
etab_pers_agees = df.loc[df["LibelleCategorieAgregat"] == cat]
count_dep = etab_pers_agees["Departement"].value_counts().reset_index()
count_dep = count_dep.loc[count_dep["Departement"] <= "95"]

# on récupère les contours des départements (malheureusement pas les Outre-Mer)
contours = load_table_from_resource_id('90b9341a-e1f7-4d75-a73c-bbc010c7feeb')
```

```python
count_dep = gpd.GeoDataFrame(
    pd.merge(
        count_dep.rename({"Departement": "code"}, axis=1),
        contours,
        on="code",
    ),
    geometry="geometry",
)
count_dep
```

```python
print(f"Nombre de {cat} par département")
count_dep.plot("count", legend=True);
```

```python
# on récupère les tranches d'âges, il faut nettoyer un peu les données
population = pd.read_excel(
    "https://www.insee.fr/fr/statistiques/fichier/2012692/TCRD_021.xlsx",
    skiprows=3,
)
population
```

```python
clean_pop = population.rename(
    {population.columns[0]: "code", population.columns[-1]: ">75"},
    axis=1,
)[["code", "Ensemble", ">75"]]
clean_pop = clean_pop.loc[clean_pop["code"] <= "95"]
clean_pop["nb_>75"] = round(clean_pop["Ensemble"] * clean_pop[">75"] / 100)
clean_pop
```

```python
final = gpd.GeoDataFrame(
    pd.merge(
        count_dep,
        clean_pop,
        on="code",
    ),
    geometry="geometry",
)
final["ratio_>75_nb_etab"] = round(final["nb_>75"] / final["count"], 2)
final
```

```python
final.plot("ratio_>75_nb_etab", legend=True);
```
