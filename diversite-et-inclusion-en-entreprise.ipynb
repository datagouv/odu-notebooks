# Défi open data : Diversité et inclusion en entreprise

Les entreprises mettent en place des politiques pour favoriser la diversité et l’inclusion dans leurs effectifs. Ces politiques peuvent être mesurées et leur efficacité évaluée à travers plusieurs documents, dont les bilans sociaux.

Depuis 1977, les entreprises sont incitées à réaliser et publier un [bilan social](https://code.travail.gouv.fr/glossaire/bilan-social). Ils sont destinés à fournir un aperçu de la situation sociale de l’entreprise, couvrant plusieurs aspects comme l’emploi, la formation, les conditions de travail, la rémunération, ou la diversité. Obligatoires pour les entreprises de plus de 300 salariés, ces bilans permettent aux parties prenantes, y compris les employés, les syndicats et les pouvoirs publics, d’avoir une vision transparente des politiques sociales d’une entreprise.

Publiés à échéances régulières (a priori tous les deux ans), **ce sont des sources d’information très utiles** pour évaluer une politique d’inclusion dans le temps. Néanmoins, les bilans sociaux sont difficilement accessibles. Il n’existe pas, à ce jour, de répertoire des bilans sociaux. De plus, les données sont souvent publiées au format PDF, ce qui les rend difficilement exploitables. Pourtant, derrière les bilans sociaux, les entreprises constituent [une base de données économique et sociale](https://fr.wikipedia.org/wiki/Base_de_donn%C3%A9es_%C3%A9conomiques_et_sociales_(France)). Ces bases de données (anonymisées) sont très peu publiées par les entreprises. EDF SA [en publie une partie](https://opendata.edf.fr/explore/?disjunctive.theme&disjunctive.publisher&disjunctive.keyword&sort=modified&refine.theme=RSE+-+Responsabilit%C3%A9+soci%C3%A9tale+des+entreprises).

L’entreprise EDF SA a une stratégie sociale reposant sur la promotion et la valorisation de la diversité de ses salariés : [toutes ses actions en faveur de la diversité et l’inclusion](https://www.edf.fr/edf-recrute/pourquoi-choisir-edf/un-employeur-responsable/une-entreprise-ouverte-a-tous) intègrent la lutte contre toute forme de discrimination. À titre d’exemple, EDF a signé en 2015 la charte LGBT de [l’Autre Cercle](https://autrecercle.org/).

**La problématique du défi**

Ce défi se concentre sur l'analyse et l'évaluation des politiques d'inclusion et de diversité d'EDF SA à travers ses bilans sociaux : 

* **Quelles sont les caractéristiques importantes et évolutions clés du bilan social d’EDF SA sur l’inclusion et la diversité au sein de l’entreprise ?**
* **Comment EDF SA se situe sur ces sujets par rapport à des entreprises comparables en France, en Europe ou même ailleurs dans le monde ?**
