# Défi open data : Tourisme en train

En France, le secteur du tourisme a émis 97 millions de tonnes de CO2 en 2022 et le transport représente 69% de cette empreinte carbone. Dans ce contexte, un trajet de 500 km en train émet [presque 10 fois moins de CO2](https://agirpourlatransition.ademe.fr/particuliers/bureau/deplacements/calculer-emissions-carbone-trajets) qu'un trajet en voiture individuelle.

Face aux enjeux climatiques et à la nécessité de réduire l'empreinte carbone des déplacements, **[le tourisme durable](https://infos.ademe.fr/societe-politiques-publiques/2024/cest-lete-quelques-pistes-pour-voyager-plus-responsable/) est devenu un enjeu important pour les touristes et les agences de tourisme locales**. Des plateformes spécialisées sur le tourisme en train ont ainsi vu le jour (à l’instar de [Mollow](https://www.mollow.eu/) ou [Hourrail](https://www.hourrail.voyage/fr)) et des offices de tourisme font désormais la promotion de voyages bas-carbone ([exemple avec Normandie Tourisme](https://www.normandie-tourisme.fr/preparer-son-sejour/voyager-sans-voiture/)).

Dans ce contexte, la SNCF travaille à développer davantage le **tourisme en train**, en rendant plus accessibles et attractives les destinations desservies par son réseau. Avec plus de **3 000 gares** réparties sur l'ensemble du territoire et un maillage dense de lignes ferroviaires, la France dispose d'un potentiel important pour développer un tourisme de proximité (et multimodal) respectueux de l’environnement. Néanmoins, **le train n'est pas toujours le premier choix des voyageurs** pour leurs déplacements touristiques (méconnaissance des destinations accessibles, manque d'informations sur les attractions touristiques à proximité des gares, difficultés à trouver des modes de déplacement sobres une fois sur place, etc.).

# Exemple de réalisation : nombre de gares exploitées par département

```python
import requests
import geopandas as gpd
```

```python
unstacked = pd.concat(
    [
        df[["CODE_LIGNE", "LIB_LIGNE", "STATUT", "C_GEO_D"]].rename({"C_GEO_D": "C_GEO"}, axis=1),
        df[["CODE_LIGNE", "LIB_LIGNE", "STATUT", "C_GEO_F"]].rename({"C_GEO_F": "C_GEO"}, axis=1),
    ],
    ignore_index=True
)
unstacked
```

```python
# on récupère les infos géographiques à partir de coordonnées
base_url = "https://geo.api.gouv.fr/communes?lat={}&lon={}"

def get_infos(coords: str):
    r = requests.get(base_url.format(*coords.split(", ")))
    if not r.ok:
        return None
    res = r.json()
    if res:
        return res[0]
    return None

unstacked["api_res"] = unstacked["C_GEO"].apply(get_infos)
# on retire les points sans infos (à l'étranger en majorité)
unstacked = unstacked.dropna(subset="api_res")
```

```python
unstacked["code_commune"], unstacked["code_dep"], unstacked["code_reg"], unstacked["pop"] = zip(
    *unstacked["api_res"].apply(
        lambda d: (d["code"], d["codeDepartement"], d["codeRegion"], d["population"]),
    ),
)
```

```python
open_stats = unstacked.loc[unstacked["STATUT"] == "Exploitée", "code_dep"].value_counts().reset_index().rename(
    {"code_dep": "code"},
    axis=1,
)
open_stats
```

```python
# on récupère les contours des départements (malheureusement pas les Outre-Mer)
contours = load_table_from_resource_id('90b9341a-e1f7-4d75-a73c-bbc010c7feeb')
contours
```

```python
moy_geo = gpd.GeoDataFrame(pd.merge(
    open_stats,
    contours,
    on="code",
    how="left",
), geometry="geometry")
moy_geo.plot("count", legend=True);
```
